And adding the newest emojis from the 2025 emoji list: [https://www.prosettings.com/emoji-list/](https://www.prosettings.com/emoji-list/)

The dictionaries can be found in ./Lists directory


```python
from Lists.abbreviations import INTERNET_SLANG
from Lists.emoticons import EMOTICONS
from Lists.emojis import EMO_UNICODE
import re
```

Testing emojis

```python
UNICODE_EMO = {v: k for k, v in EMO_UNICODE.items()}
```

```python
def convert_emojis(text):
    for emot in UNICODE_EMO:
        text = re.sub(r'('+emot+')', " ".join(UNICODE_EMO[emot].replace(",","").replace(":","").split()), text)
    text = text.replace('_', ' ')
    return text
```

```python
text = "Kids Love😘❤"
converted = convert_emojis(text)
print(converted)
```

Testing emoticons

```python
def convert_emoticons(text):
    for emot in EMOTICONS:
        text = re.sub(u'('+emot+')', " ".join(EMOTICONS[emot].replace(",","").split()), text)
    return text
```

```python
text = "I am so happy :) but yesterday I was :( and a little crying :'(."

converted = convert_emoticons(text)
print(converted)
```

Testing abbreviations

```python
def convert_abbreviations(text):
    for abbr in INTERNET_SLANG:
        pattern = re.compile(r'\b' + re.escape(abbr) + r'\b', re.IGNORECASE)
        replacement = " ".join(INTERNET_SLANG[abbr].replace(",", "").split())
        text = pattern.sub(replacement, text)
    return text
```

```python
text = "FYI, I am AFK but brb. LOL!"

print(convert_abbreviations(text))
```
