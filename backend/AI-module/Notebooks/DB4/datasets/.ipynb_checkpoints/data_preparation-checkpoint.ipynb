```python
dfs = [df_1, df_2, df_3, df_4, df_5, df_6, df_7, df_8]
df = pd.concat(dfs, ignore_index=True)
```

```python
df = df[df['type'] != 'cyberbullying']
df = df.sample(n=101476, random_state=42)
print(df['type'].value_counts())
df.to_csv('./all_datasets_not_cyberbullying.csv',index=False)
```

```python
df_OLID = pd.read_csv('./datasets/OLID.csv')

df_OLID = df_OLID.drop(columns=['label_C', 'label_B'])

df_OLID = df_OLID.rename(columns={'label_A': 'type'})
df_OLID['type'] = df_OLID['type'].replace({'OFF': 'cyberbullying', 'NOT': 'not_cyberbullying'})
df_OLID['label'] = df_OLID['type'].apply(lambda x: 1 if x == 'cyberbullying' else 0)
df_OLID = df_OLID[df_OLID['label'] == 1].reset_index(drop=True)

df_OLID.head()
```

```python
print(df_OLID['type'].value_counts())
```

```python
df_cyber = pd.read_csv('./datasets/cyberbullying_tweets.csv')
df_cyber = df_cyber.rename(columns={'tweet_text': 'text','cyberbullying_type':'type'})

df_cyber['type'] = df_cyber['type'].apply(lambda x: 'cyberbullying' if x != 'not_cyberbullying' else 'not_cyberbullying')
df_cyber['label'] = df_cyber['type'].apply(lambda x: 1 if x == 'cyberbullying' else 0)
df_cyber = df_cyber[df_cyber['label'] == 1].reset_index(drop=True)
df_cyber.head()
```

```python
print(df_cyber['type'].value_counts())
```

```python
df_loyola = pd.read_csv('./datasets/dataset_loyola_university.csv')
df_loyola = df_loyola[['Comment', 'Type']]

df_loyola = df_loyola.rename(columns={'Comment': 'text','Type':'label'})
df_loyola['label'] = df_loyola['label'].fillna(0).astype(int)

df_loyola['type'] = df_loyola['label'].apply(lambda x: 'cyberbullying' if x == 1 else 'not_cyberbullying')
df_loyola = df_loyola[df_loyola['label'] == 1].reset_index(drop=True)
df_loyola.head()
```

```python
print(df_loyola['type'].value_counts())
```

```python
df = df[['text', 'type', 'label']]
df_OLID = df_OLID[['text', 'type', 'label']]
df_loyola = df_loyola[['text', 'type', 'label']]
df_cyber = df_cyber[['text', 'type', 'label']]
```

```python
df = pd.concat([df, df_OLID, df_loyola, df_cyber], ignore_index=True)
```

```python
print(df.shape)
print(df['type'].value_counts())
```

```python
#undersampling
df_cyberbullying = df[df['type'] == 'cyberbullying']
df_not_cyberbullying = df[df['type'] == 'not_cyberbullying']

df_not_cyberbullying_sampled = df_not_cyberbullying.sample(n=len(df_cyberbullying), random_state=42)

df_balanced = pd.concat([df_cyberbullying, df_not_cyberbullying_sampled], ignore_index=True)

df_balanced = df_balanced.sample(frac=1, random_state=42).reset_index(drop=True)

print(df_balanced['type'].value_counts())
print(df_balanced.shape)
```

```python
df_balanced.to_csv('./all_datasets.csv',index=False)
```
