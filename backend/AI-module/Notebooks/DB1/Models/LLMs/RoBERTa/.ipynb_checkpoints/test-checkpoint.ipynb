```python
import torch
import torch.nn as nn
from transformers import RobertaModel, RobertaTokenizer
from transformers import RobertaConfig
import numpy as np
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
class Roberta_Classifier(nn.Module):
    def __init__(self,
                 freeze_bert=False,
                 bert_dropout=0.3,
                 classifier_dropout=0.3):
        super(Roberta_Classifier, self).__init__()

        n_input = 768
        n_hidden = 50
        n_output = 5

        config = RobertaConfig.from_pretrained(
            "roberta-base",
            hidden_dropout_prob=bert_dropout,
            attention_probs_dropout_prob=bert_dropout
        )
        self.roberta = RobertaModel.from_pretrained(
            "roberta-base",
            config=config
        )

        self.classifier = nn.Sequential(
            nn.Dropout(classifier_dropout),
            nn.Linear(n_input, n_hidden),
            nn.LeakyReLU(),
            nn.Dropout(classifier_dropout),
            nn.Linear(n_hidden, n_output)
        )

        if freeze_bert:
            for param in self.roberta.parameters():
                param.requires_grad = False

    def forward(self, input_ids, attention_mask):
        outputs = self.roberta(
            input_ids=input_ids,
            attention_mask=attention_mask
        )

        cls_emb = outputs.last_hidden_state[:, 0, :]
        logits = self.classifier(cls_emb)
        return logits
```

```python
model = Roberta_Classifier(freeze_bert=False).to(device)
state = torch.load("roberta_cyberbullying_model.pth", map_location=device)
model.load_state_dict(state, strict=False)
model.eval()
```

```python
tokenizer = RobertaTokenizer.from_pretrained("roberta-base")
```

```python
import torch
import torch.nn.functional as F

def classify_text(text, 
                  model, 
                  tokenizer, 
                  max_length=256, 
                  device=device):
    """
    Primește un text (string), un model Roberta_Classifier și tokenizer-ul aferent,
    și returnează (predicted_label, probabilities_list).
    """
    model.eval()
    model.to(device)

    # Tokenizare „batch‐style” și returnare PyTorch tensors
    encoded = tokenizer(
        text,
        add_special_tokens=True,
        max_length=max_length,
        padding='max_length',
        truncation=True,
        return_tensors='pt'
    )
    input_ids = encoded['input_ids'].to(device)
    attention_mask = encoded['attention_mask'].to(device)

    with torch.no_grad():
        logits = model(input_ids, attention_mask)
        probs  = F.softmax(logits, dim=-1).cpu().tolist()[0]
        pred_idx = int(torch.argmax(logits, dim=-1).cpu().item())

    # Etichetele tale (sunt 5 ieșiri în model)
    class_labels = [
        "religion",
        "age",
        "ethnicity",
        "gender",
        "other_cyberbullying"
    ]

    return class_labels[pred_idx], probs
```

```python
class_labels = [
    "religion",
    "age",
    "ethnicity",
    "gender",
    "other_cyberbullying"
]
```

```python
texts = [
    "rape is real..zvasiyana nema jokes about being drunk or being gay or being lesbian...rape is not ones choice or wish..thtz where the sensitivity is coming from",
    "same girl that bullied me everyday in high school now posts motivational shit on her story like fuck offffff",
    "Dumb ass nigger not knowing how to drive fuck you I drive swell"
]

for text in texts:
    label, probs = classify_text(
        text,
        model=model,
        tokenizer=tokenizer,
        max_length=256,
        device=device
    )
    print(f"\nText: {text!r}")
    print(f"  Predicted label : {label}")
    print("  Probabilities:")
    for cls, p in zip(class_labels, probs):
        print(f"    {cls:20s}: {p:.4f}")
```
