```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import Normalizer
import pandas as pd
import joblib
```

```python
df = pd.read_csv('../../cyberbullying_tweets_classifiers.csv')
df.head()
```

```python
df['word_count'] = df['text'].astype(str).apply(lambda x: len(x.split()))
df = df[df['word_count'] > 1]
```

```python
tfidf = TfidfVectorizer(
    ngram_range=(1, 2),
    max_features=10000,
    sublinear_tf=True,
    stop_words='english',
    min_df=2
)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df['text'], df['label'], test_size=0.2, random_state=42, stratify=df['label']
)
X_train_tfidf = tfidf.fit_transform(X_train.astype('U')).toarray()
X_test_tfidf = tfidf.transform(X_test.astype('U')).toarray()

normalizer = Normalizer()
X_train_norm = normalizer.fit_transform(X_train_tfidf)
X_test_norm = normalizer.transform(X_test_tfidf)
```

```python
joblib.dump(X_train_tfidf, './Data/X_train_tfidf_1.pkl')
joblib.dump(X_test_tfidf, './Data/X_test_tfidf_1.pkl')
joblib.dump(y_train, './Data/y_train_tfidf_1.pkl')
joblib.dump(y_test, './Data/y_test_tfidf_1.pkl')
```

```python
joblib.dump(X_train_norm, './Data/X_train_tfidf_2.pkl')
joblib.dump(X_test_norm, './Data/X_test_tfidf_2.pkl')
joblib.dump(y_train, './Data/y_train_tfidf_2.pkl')
joblib.dump(y_test, './Data/y_test_tfidf_2.pkl')
joblib.dump(tfidf, './Data/tfidf_vectorizer2.pkl')
```
