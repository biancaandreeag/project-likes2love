```python
dfs = [df_1, df_2, df_3, df_4, df_5]
df = pd.concat(dfs, ignore_index=True)
df.info()
```

```python
df = df.rename(columns={
    "CB Comment: Whether comment overall was a CB comment or not (1= CB 0= No CB)": "Type"
})

df.drop(columns=[
    "Other Phenomena: Other aspects that might be applicable to the CB comment",
    "Unnamed: 1",
    "Role: Purpose of the CB comment"
], inplace=True)


df["Type"] = df["Type"].map({
    0.0: "not_cyberbullying",
    1.0: "cyberbullying"
})

df.head()
```

```python
import matplotlib.pyplot as plt

counts = df["Type"].value_counts()

# Grafic
counts.plot(kind='bar', color=['skyblue', 'salmon'])
plt.title("Distribuția comentariilor după tip")
plt.xlabel("Tip comentariu")
plt.ylabel("Număr comentarii")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
cols_to_check = [
    "1= Sexual; but NOT sexual/gender identity (e.g. \"whore\") 0= N/A",
    "1= Sexual orientation/gender identity (e.g. \"gay\") 0= N/A",
    "1 = Physical Appearance 0= N/A",
    "1= Race/Ethnicity 0= N/A",
    "1= Intellectual 0= N/A",
    "1= Religious 0= N/A",
    "1 = General Hate (F* Off, Go Die, etc) 0= N/A",
    "1= Insult /Attack 0= N/A",
    "1= Defense of Another 2= Defense of Self 0= N/A",
    "1= Depression 0= N/A",
    "1= Suicide 0= N/A",
    "1= Stress (or anxiety) 0= N/A",
    "1= Discrimination 0= N/A"
]

# Dicționar pentru rezultate
results = {}

# Parcurge fiecare coloană și contorizează valorile
for col in cols_to_check:
    value_counts = df[col].value_counts()
    results[col] = {
        "0.0": int(value_counts.get(0.0, 0)),
        "1.0": int(value_counts.get(1.0, 0))
    }

# Convertește în DataFrame pentru afișare frumoasă
summary_df = pd.DataFrame(results).T
summary_df.columns = ['Count 0.0', 'Count 1.0']

# Afișează rezultatul
print(summary_df)
```
