```python
import pandas as pd

xls = pd.ExcelFile('./CB IG Compiled Coding Sets/CB IG Compiled Coding Set 1 - 2020 (Revised 2.0).xlsx')  
df_list = [xls.parse(sheet) for sheet in xls.sheet_names]
df_1 = pd.concat(df_list, ignore_index=True)

df_1 = df_1.drop(columns=['Commenter Username'])
df_1 = df_1.drop(columns=['Comment Number'])
df_1 = df_1.fillna(0.0)
```

```python
import pandas as pd

xls = pd.ExcelFile('./CB IG Compiled Coding Sets/CB IG Compiled Coding Set 5 - 2020 (Revised 2.0).xlsx')  
df_list = [xls.parse(sheet) for sheet in xls.sheet_names]
df_5 = pd.concat(df_list, ignore_index=True)

df_5 = df_5.fillna(0.0)
```

```python
df_5.head()
```

```python
dfs = [df_1, df_2, df_3, df_4, df_5]
df = pd.concat(dfs, ignore_index=True)
df.info()
```

```python
df = df.rename(columns={
    "CB Comment: Whether comment overall was a CB comment or not (1= CB 0= No CB)": "Type"
})

df.drop(columns=[
    "Other Phenomena: Other aspects that might be applicable to the CB comment",
    "Unnamed: 1",
    "Role: Purpose of the CB comment"
], inplace=True)


df["Type"] = df["Type"].map({
    0.0: "not_cyberbullying",
    1.0: "cyberbullying"
})

df.head()
```
