```python
import re
import nltk
import pandas as pd
import seaborn as sns
from wordcloud import STOPWORDS
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from nltk.probability import FreqDist
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords as nltk_stopwords

stopwords = set(nltk_stopwords.words('english'))
```

```python
df = df.rename(columns={
    '1= Sexual; but NOT sexual/gender identity (e.g. "whore") 0= N/A': 'sexual',
    '1= Sexual orientation/gender identity (e.g. "gay") 0= N/A': 'gender_identity',
    '1 = Physical Appearance 0= N/A': 'appearance',
    '1= Race/Ethnicity 0= N/A': 'race_ethnicity',
    '1= Intellectual 0= N/A': 'intellectual',
    '1= Religious 0= N/A': 'religious',
    '1 = General Hate (F* Off, Go Die, etc) 0= N/A': 'general_hate',
    '1= Insult /Attack 0= N/A': 'insult',
    '1= Defense of Another 2= Defense of Self 0= N/A': 'defense',
    '1= Depression 0= N/A': 'depression',
    '1= Suicide 0= N/A': 'suicide',
    '1= Stress (or anxiety) 0= N/A': 'stress',
    '1= Discrimination 0= N/A': 'discrimination'
})
df = df.drop(columns=['stress', 'depression'])
```

```python
all_text = ' '.join(cyber_df['text'].astype(str).tolist()).lower()

tokens = word_tokenize(all_text, language='english')

filtered_tokens = [
    word for word in tokens 
    if len(word) > 3 
    and word not in stopwords 
]

freq_dist = FreqDist(filtered_tokens)

num_top_words = 40
top_words = freq_dist.most_common(num_top_words)

words, frequencies = zip(*top_words)

plt.figure(figsize=(10, 12))
plt.barh(words[::-1], frequencies[::-1], color='#90ee90', edgecolor='black', linewidth=1.2, height=0.6)
plt.xlabel('Frequency')
plt.ylabel('Words')
plt.title('Top Words by Frequency Text Column')
plt.tight_layout()
plt.show()
```

```python
data = df 

labels = data['type'].unique()

for label in labels:
    filtered_data = data[data['type'] == label]
    
    all_posts = ' '.join(filtered_data['text'].astype(str).tolist()).lower()
    
    all_posts = re.sub(r'[^\w\s]', '', all_posts)
    
    tokens = word_tokenize(all_posts, language='english')
    
    filtered_tokens = [word for word in tokens if len(word) > 3 and word not in stopwords and word != 'user']
    
    freq_dist = FreqDist(filtered_tokens)
    
    num_top_words = 10 
    top_words = freq_dist.most_common(num_top_words)
    
    if not top_words:
        print(f"No words found for label {label}")
        continue
    
    words, frequencies = zip(*top_words)
    
    plt.figure(figsize=(10, 6))
    plt.barh(words[::-1], frequencies[::-1], color='#90ee90', edgecolor='black')  
    plt.xlabel('Frequency')
    plt.ylabel('Words')
    plt.title(f'Most Frequent Words for label {label}')
    plt.tight_layout()
    plt.show()
```

```python
for category in binary_columns:
    filtered_data = df[df[category] == 1]
    
    if filtered_data.empty:
        print(f"No comments found for category '{category}'")
        continue
    
    all_text = ' '.join(filtered_data['text'].astype(str).tolist()).lower()
    all_text = re.sub(r'[^\w\s]', '', all_text)
    
    tokens = word_tokenize(all_text, language='english')
    filtered_tokens = [w for w in tokens if len(w) > 3 and w not in stopwords and w != 'user']

    freq_dist = FreqDist(filtered_tokens)
    
    num_top_words = 10
    top_words = freq_dist.most_common(num_top_words)
    
    if not top_words:
        print(f"No words found for category '{category}'")
        continue
    
    words, frequencies = zip(*top_words)
    
    plt.figure(figsize=(10, max(5, num_top_words * 0.5)))
    plt.barh(words[::-1], frequencies[::-1], color='#90ee90', edgecolor='black')
    plt.xlabel('Frequency')
    plt.ylabel('Words')
    plt.title(f'Most Frequent Words for Category: {category}')
    plt.tight_layout()
    plt.show()
```
